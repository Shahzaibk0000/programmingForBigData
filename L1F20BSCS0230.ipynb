```python
# Qustion#1 Find the total number of patients of each disease (diagnosis) and plot in histogram.
```

```python
import pandas as pd
import numpy as py
import matplotlib as plt
from matplotlib import pyplot
```

```python
data = pd.read_csv('jsrt_metadata.csv')
```

```python
count = data['diagnosis'].value_counts()
print(count)
```

```python
data['diagnosis']=data['diagnosis'].astype('str')
plt.pyplot.hist(data['diagnosis'],color='yellow',ec='black',orientation='horizontal')
plt.pyplot.title('Number of People Diagnosis')
plt.pyplot.xlabel('Frequency')
plt.pyplot.ylabel('Category')
```

```python
# Question#2 Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.
```

```python
plt.pyplot.pie(data[['diagnosis','state']].value_counts(),autopct='%1.1f%%')
```

```python
# Question#3 Find and plot via pia chart the total number of patients of each disease (diagnosis) with respect to the gender(Female, 
# Male)
```

```python
plt.pyplot.pie(data[['diagnosis','gender']].value_counts(),autopct='%1.1f%%')
```
